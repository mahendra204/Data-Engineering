```python
import pandas as pd
import numpy as np
```

```python
customers= pd.read_csv("customers.csv")
orders =pd.read_csv("orders.csv")
orderitems = pd.read_csv("orderitems.csv")
payments=pd.read_csv("payments.csv")
products = pd.read_csv("Products.csv")
```

# Exploratory Data Analysis (EDA) + Data Engineering

```python
customers
```

```python
no_of_custs=customers['customer_id'].nunique()
no_of_custs
```

```python
no_of_states=customers['customer_state'].nunique()
no_of_states
```

```python
no_of_cities=customers['customer_city'].nunique()
no_of_cities
```

```python
state_wise_no_of_custs=customers.groupby( 'customer_state')['customer_id'].count().reset_index(name='no_of_customers')
state_wise_no_of_custs=state_wise_no_of_custs.sort_values(by='no_of_customers', ascending = False)
state_wise_no_of_custs.head()
```

```python
statewise_no_of_cities= customers.groupby('customer_state')['customer_city'].nunique().reset_index(name='no_of_cities')
statewise_no_of_cities= statewise_no_of_cities.sort_values(by='no_of_cities', ascending = False)
statewise_no_of_cities
```

```python
statewise_no_of_custs_cities = pd.merge(state_wise_no_of_custs,statewise_no_of_cities, on='customer_state', how = 'inner')
statewise_no_of_custs_cities['customers/city']=statewise_no_of_custs_cities['no_of_customers']/statewise_no_of_custs_cities['no_of_cities']
statewise_no_of_custs_cities.sort_values(by='no_of_customers', ascending = False, inplace=True)
statewise_no_of_custs_cities
```

```python
orders
```

```python
no_of_orders = orders['order_id'].nunique()
no_of_orders
```

```python
order_statuses=orders['order_status'].unique()
order_statuses
```

```python
no_orders_status_wise=orders.groupby('order_status')['order_id'].nunique().reset_index(name='no_of_orders')
no_orders_status_wise['percentage_orders']=no_orders_status_wise['no_of_orders']*100/no_of_orders
no_orders_status_wise.round(2).sort_values(by='percentage_orders', ascending = False)
```

```python
orders_timings=orders[['order_status','order_purchase_timestamp','order_delivered_timestamp','order_estimated_delivery_date']]
orders_timings['order_purchase_timestamp']=pd.to_datetime(orders_timings['order_purchase_timestamp'])
orders_timings['order_delivered_timestamp']=pd.to_datetime(orders_timings['order_delivered_timestamp'])
orders_timings['order_estimated_delivery_date']=pd.to_datetime(orders_timings['order_estimated_delivery_date'])
orders_timings
```

```python
orders_timings['estimated_delivery_no_of_days']=orders_timings['order_estimated_delivery_date']-orders_timings['order_purchase_timestamp']
orders_timings['actudal_delivery_no_of_days'] = orders_timings['order_delivered_timestamp']-orders_timings['order_purchase_timestamp']
orders_timings.sort_values(by='actudal_delivery_no_of_days', ascending =False)
```

```python
non=orders_timings[orders_timings['order_status']!='delivered']
non
```

```python
product_details=orderitems[['product_id', 'price']].drop_duplicates()
#products['price']=orderitems['price']
product_details= product_details.sort_values(by='product_id', ascending=False)
#products.shape
product_details
```

```python
product_charges_order_wise=orderitems[['order_id', 'product_id', 'price', 'shipping_charges']]
product_charges_order_wise['total_amount']=product_charges_order_wise['price']+product_charges_order_wise['shipping_charges']
product_charges_order_wise=product_charges_order_wise.groupby(['order_id', 'product_id', 'price','shipping_charges'])['total_amount'].sum().reset_index(name='Total_amount')
product_charges_order_wise.sort_values(by='Total_amount', ascending=False, inplace=True)
product_charges_order_wise
```

```python
product_count=products['product_id'].nunique()
product_count
```

```python
product_categories=products['product_category_name'].nunique()
product_categories
```

```python
product_category_names=products['product_category_name'].unique()
product_category_names
```

```python
category_wise_no_of_products=products.groupby('product_category_name')['product_id'].nunique().reset_index(name='no_of_products')
category_wise_no_of_products=category_wise_no_of_products.sort_values(by='no_of_products', ascending =False)
category_wise_no_of_products.head().reset_index()
```

```python
top_5_heaviest_products = (
    products.drop_duplicates(subset='product_id')  # Ensure distinct product IDs
    .sort_values(by='product_weight_g', ascending=False)  # Sort by weight in descending order
    .groupby('product_category_name', group_keys=False)  # Group by category
    .apply(lambda x: x.head(5))  # Get top 5 products per category
)
```

```python
top_5_heaviest_products.head()
```

```python
top_5_heaviest_products.tail()
```

```python
products['volume']=products['product_length_cm']*products['product_height_cm']*products['product_width_cm']
products
```

```python
top_5_moresized_products = (
    products.drop_duplicates(subset='product_id')  # Ensure distinct product IDs
    .sort_values(by='volume', ascending=False)  # Sort by weight in descending order
    .groupby('product_category_name', group_keys=False)  # Group by category
    .apply(lambda x: x.head(5))  # Get top 5 products per category
)
```

```python
top_5_moresized_products.head()
```

```python
top_5_moresized_products.tail()
```

```python
payments
```

```python
payment_types=payments['payment_type'].unique()
payment_types
```

```python
payment_types=payments['payment_type'].nunique()
payment_types
```

```python
payment_type_orders=payments.groupby('payment_type')['order_id'].nunique().reset_index(name='no_of_orders')
payment_type_orders['percentage_orders']=payment_type_orders['no_of_orders']*100/no_of_orders
payment_type_orders.sort_values(by='percentage_orders', ascending=False, inplace=True)
payment_type_orders
```

```python
payment_installments_no_of_orders=payments.groupby('payment_installments')['order_id'].nunique().reset_index(name='no_orders')
payment_installments_no_of_orders['percent_orders']=payment_installments_no_of_orders['no_orders']*100/no_of_orders
payment_installments_no_of_orders=payment_installments_no_of_orders.sort_values(by='no_orders', ascending=False)
payment_installments_no_of_orders
```

```python
payment_orders_payment=payments[['order_id', 'payment_value','payment_installments']]
payment_orders_payment=payment_orders_payment.sort_values(by='payment_value', ascending=False)
payment_orders_payment
```

```python
totalpayment=payments['payment_value'].sum()
totalpayment
```

```python
paymenttype_amount=payments.groupby('payment_type')['payment_value'].sum().reset_index(name='total_payment')
paymenttype_amount
```

```python
customer_payments=payment_orders_payment.merge(orders, on='order_id', how='inner')
customer_payments=customer_payments[['customer_id', 'order_id', 'payment_value','payment_installments']]
customer_payments.sort_values(by='payment_value', ascending=False)
customer_payments
```

```python
customer_payment_types=payments.merge(orders, on='order_id', how='inner')
customer_payment_type=customer_payment_types[['customer_id', 'order_id', 'payment_type']]
customer_payment_type
```

```python
payment_type_custs=customer_payment_type.groupby('payment_type')['customer_id'].nunique().reset_index(name='no_of_customers')
payment_type_custs
```

```python
paymenttype_amount.merge(payment_type_custs, on='payment_type', how='inner')
paymenttype_amount
```

```python
products_orderwise=products.merge(orderitems, on='product_id', how='inner')
products_orderwise= products_orderwise[['product_id','order_id','product_category_name']]
cat_wise_payment_value= products_orderwise.merge(payments, on='order_id', how='inner')
cat_wise_payment_value
```

```python
cat_wise_payment_values=cat_wise_payment_value[['order_id', 'product_id','product_category_name', 'payment_value']]
cat_wise_payment_values
```

```python
category_wise_payment_values=cat_wise_payment_values.groupby('product_category_name')['payment_value'].sum().reset_index(name='category_wise_payment_value')
category_wise_payment_values.sort_values(by='category_wise_payment_value', ascending=False)
category_wise_payment_values
```
